## FinanceDataReader 사용법

```pandas-datareader```의 불안전성을 보완하고자 만든 API로
+ 한국 주식 가격
+ 미국 주식 가격
+ 지수
+ 환율
+ 암호 화폐 가격
+ 종목 리스트
<br>
등을 제공한다.

````python
# 한국 거래소에 상장된 모든 종목 리스트 불러오기
import FinanceDataReader as fdr

df_krx = fdr.StockListing('KRX')
print(len(df_krx))
df_krx.head()
````

````python
# 가격 데이터 가져오기 - GS글로벌 -

df = fdr.DataReader('001250', '2018') # GS글로벌, 2018년
print(df.head(10))
print(df['Close'].plot())
````

````python
# 가격 데이터 가져오기 - AAPL -

df = fdr.DataReader('AAPL', '2017') # GS글로벌, 2018년
print(df.head(10))
print(df['Close'].plot())
````

이 외에
+ 각종 지수 : KS11(코스피 지수), KQ11(코스닥 지수), DJI(다우 지수), IXIC(나스닥 지수), US500(S&P500)
+ 환율 데이터 : USD/KRX(원달러 환율), USD/EUR(달러당 유로화 환율), CNY/KRW(위안화 원화 환율)
+ 암호 화폐 가격 : BTC/USD(비트코인 달러 가격, 비트파이넥스), BTC/KRW(비트코인 원화 가격, 빗썸)
<br>
등 다양한 값을 호출 가능하다.
